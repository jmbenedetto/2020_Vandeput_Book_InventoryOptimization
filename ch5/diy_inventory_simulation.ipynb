# Overview


The chapter relax the determinist constraint over the demand. Demand now is stochastic, although limited to a normal distribution.


# Packages


```python
import numpy as np
import pandas as pd
from scipy.stats import norm
```

# Parameters


```python
time = 200
d_mu = 100
d_std = 25
L = 4
R = 1
alpha = 0.95
```

```python
# This is a placeholder for your code
```

# Script


```python
# define inventory variables
d = np.maximum(np.random.normal(d_mu, d_std, time).round(0).astype(int), 0)
z = norm.ppf(alpha)
x_std = np.sqrt(L + R) * d_std
Ss = np.round(x_std * z).astype(int)
Cs = 1 / 2 * d_mu * R
Is = d_mu * L
S = Ss + 2 * Cs + Is

# define arrays to store results
hand = np.zeros(time, dtype=int)
transit = np.zeros((time, L + 1), dtype=int)
stockout_period = np.full(time, False, dtype=bool)
stockout_cycle = []

# initialize arrays
hand[0] = S - d[0]
transit[1, -1] = d[0]
```

```python
print("time: ", time)
print("d_mu: ", d_mu)
print("d_std: ", d_std)
print("L: ", L)
print("R: ", R)
print("alpha: ", alpha)
print("d: ", d)
print("z: ", z)
print("x_std: ", x_std)
print("Ss: ", Ss)
print("Cs: ", Cs)
print("Is: ", Is)
print("S: ", S)
print("hand: ", hand)
print("transit: ", transit)
print("stockout_period: ", stockout_period)
print("stockout_cycle: ", stockout_cycle)
```
